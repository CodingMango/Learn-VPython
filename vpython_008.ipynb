```python
from vpython import *
import numpy as np
canvas()
orb = sphere()

while True:
    #cycle-1
    for red in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(red,0,0)
    for green in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(1,green,0)
    for blue in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(1,1,blue)
    for red in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(red,1,1)
    for green in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(0,green,1)
    for blue in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(0,0,blue)
    #cycle-2
    for green in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(0,green,0)
    for blue in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(0,1,blue)
    for red in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(red,1,1)
    for green in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(1,green,1)
    for blue in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(1,0,blue)
    for red in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(red,0,0)
    #cycle-3
    for blue in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(0,0,blue)
    for red in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(red,0,1)
    for green in np.linspace(0,1,100):
        rate(100)
        orb.color=vector(1,green,1)
    for blue in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(1,1,blue)
    for red in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(red,1,0)
    for green in np.linspace(1,0,100):
        rate(100)
        orb.color=vector(0,green,0)
```

```python
#Better program
from vpython import *
canvas()
orb=sphere()  #NOTE: the opacity of the orb is not always equal (brightness is varying)
rChan=0
bChan=0
gChan=0
rInc=.001
bInc=.002
gInc=.0015
while True:
    rate(200)
    rChan+=rInc
    bChan+=bInc
    gChan+=gInc
    orb.color=vector(rChan,gChan,bChan)
    if rChan<=0 or rChan>=1:
        rInc*=-1
    if bChan<=0 or bChan>=1:
        bInc*=-1
    if gChan<=0 or gChan>=1:
        gInc*=-1
```

```python
#Colorful Orb with continuous brightness (program-1)
#Hint: make rChan+bChan+gChan always equals 1
from vpython import *
canvas()
orb=sphere()
rChan=1
bChan=0
gChan=1
rInc=.001
bInc=.001
gInc=.001
while True:
    for i in range(0,1000): #red is constant
        rate(100)
        gChan-=gInc
        bChan+=bInc
        orb.color=vector(rChan,gChan,bChan)
    for i in range(0,1000): #blue is constant
        rate(100)
        rChan-=rInc
        gChan+=gInc
        orb.color=vector(rChan,gChan,bChan)
    for i in range(0,1000): #green is constant
        rate(100)
        rChan+=rInc
        bChan-=bInc
        orb.color=vector(rChan,gChan,bChan)

#NOTE: if you print(rChan+bChan+gChan) anywhere in the loop you will get approx 2 (may with some rounding error)
```

```python
#Colorful Orb with continuous brightness (program-2)*
#same as above program but with different approach
from vpython import *
canvas()
orb=sphere()
rChan=1
bChan=0
gChan=1
rInc=.001
bInc=.001
gInc=-.001
while True:
    rate(100)
    rChan+=rInc
    bChan+=bInc
    gChan+=gInc
    
    if rChan<=1:
        rApply=rChan
    if rChan>1:
        rApply=1

    if bChan<=1:
        bApply=bChan
    if bChan>1:
        bApply=1

    if gChan<=1:
        gApply=gChan
    if gChan>1:
        gApply=1
    orb.color=vector(rApply,gApply,bApply)
    
    if rChan<=0 or rChan>=1.5 :
        rInc*=-1
    if bChan<=0 or bChan>=1.5 :
        bInc*=-1
    if gChan<=0 or gChan>=1.5 :
        gInc*=-1
#NOTE: if you print((rApply+gApply+bApply) anywhere in the loop you will get approx 2 (may with some rounding error)
```
