#  Thousand Cubes Visualisation 

This notebook demonstrates how to create a 3D grid of cubes using VPython and animate them.

## What you'll learn
- Creating 3D objects using `box`
- Using nested loops for spatial arrangement
- Applying color gradients
- Animating objects using rotation

### ⚠ NOTE: "Next step" will only execute if you run the "previous step".
> Because the imports, variable declaration, etc., are written in separate cells to teach you the steps.


---

## Step 1: Import VPython

We start by importing the VPython library.


````python
from vpython import *
````

---

## Step 2: Initialize Variables

- `bigCube`: stores all cube objects  
- Color variables (`redColor`, `greenColor`, `blueColor`) control gradients  
- `colorInc`: increment step for smooth color transitions


````python
bigCube = []
redColor = 0
greenColor = 0
blueColor = 0
colorInc = 0.1
````

---

## Step 3: Create a 3D Grid of Cubes

We use three nested loops to position cubes in 3D space.  
Each cube gets a slightly different color to create a gradient effect.


````python
canvas()
for xPos in range(-10,10,2):
    greenColor = 0
    for yPos in range(-10,10,2):
        redColor = 0
        for zPos in range (-10,10,2):
            cube = box(size = vector(1,1,1),
                       pos = vector(xPos,yPos,zPos),
                       color = vector(redColor, greenColor, blueColor)
                       )
            bigCube.append(cube)
            redColor += colorInc
        greenColor += colorInc
    blueColor += colorInc
````

---

## Step 4: Animate the Cubes

We continuously rotate each cube around its own position to create a dynamic animation.


```
rAngle=0.01
while True:
    rate(500)
    for cube in bigCube:
        cube.rotate(angle = rAngle , axis = vector(0,1,0), origin = cube.pos )
```

- We add this code block at the end to rotate the cubes.

---

# step 5: Combine all the above 👇


````python
from vpython import *
canvas()
bigCube = []
redColor = 0
greenColor = 0
blueColor = 0
colorInc = 0.1   
for xPos in range(-10,10,2):
    greenColor = 0
    for yPos in range(-10,10,2):
        redColor = 0
        for zPos in range (-10,10,2):
            cube = box(size = vector(1,1,1),
                       pos = vector(xPos,yPos,zPos),
                       color = vector(redColor, greenColor, blueColor)
                       )
            bigCube.append(cube)
            redColor += colorInc
        greenColor += colorInc
    blueColor += colorInc

rAngle=0.01
while True:
    rate(500)
    for cube in bigCube:
        cube.rotate(angle = rAngle , axis = vector(0,1,0), origin = cube.pos )
````
